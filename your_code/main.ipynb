#### We want to know the information about the mean salary and number of employees in each department and for each job title.

```python
# your answer here
# indexes = Department, Title
# values = mean salary, n employees
df2 = employee.pivot_table(index= ['Department','Title'], values = ['Salary', 'Name'], aggfunc = {'Salary':'mean','Name':pd.Series.nunique}).reset_index()
df2.columns = ['Department', 'Title', 'No of Employee', 'Average Salary']
df2.head(10)
```

#### We want to know for each department and for each job title the Salary's median, minimum, maximum and standard deviation. Comment your results.

```python
# your answer here
# indexes = department, title
# values = Salary
df3 = employee.pivot_table(index = ['Department', 'Title'], values = ['Salary'], aggfunc = ['median', 'min', 'max', 'std']).reset_index()
df3.columns = ['Department', 'Title', 'MedianSalary', 'MinSalary', 'MaxSalary', 'StdSalary']
df3.head(10)
```

#### The stake holders want to know for each department the number of employees and how much money is spend on salaries. Could you provide that information? 

```python
# your answer here
# indexes = department
# values = no employees, sum of saleries

df4 = employee.pivot_table(index = ['Department'], values = ['Name', 'Salary'], aggfunc = {'Name':pd.Series.nunique, 'Salary':sum}).reset_index()
df4.columns = ['Department','No_employees', 'Salary_spent']
df4.head()
```

#### For each Department and Title, we want to know the the total years of work experience and the mean salary.

```python
# your answer here
# index = Department, Title
# values = total years, mean salary

df5 = employee.pivot_table(index = ['Department', 'Title'], values = ['Years', 'Salary'], aggfunc = {'Years':sum, 'Salary':np.mean}).reset_index()
df5.columns = ['Department','Title', 'Mean_salary', 'Total_work_experience']
df5.head()
```

```python
def mean_no_max(values):
    values = sorted(values)
    values = values[:len(values)-1]
    mean = np.mean(values)
    return mean

df_mean_no_max = employee.pivot_table(index=['Department'],values= ['Salary'], aggfunc = [mean_no_max]).reset_index()
df_mean_no_max
```
